```python
#install and then import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[-3,5,7]
y=[10,2,5]
```

```python
fig=plt.figure(figsize=(8,3))
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(-3,8)
plt.xlabel('Sales',color='blue')
plt.ylabel('product')
plt.title('Line Plot')
plt.suptitle('Sales Comparison',size=20,y=1.1);
```

```python
fig,ax=plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax[0,0].set_title('Line')
ax[0,0].plot(x,y, color='Green')
ax[1,0].set_title('Scatter Plot')
ax[1,0].scatter(x,y)
ax[0,1].set_title('Piechart')
ax[0,1].pie(y)
ax[1,1].set_title('Histogram')
ax[1,1].hist(y,color='yellow')
plt.show()
```

SEABORN

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

Univariate--"one variable" data visualization

```python
#strip plot
sns.stripplot(x='sex',y='tip',data=tips,jitter=True)
plt.title('Strip Plot')
plt.xlabel('Gender')
plt.ylabel('Tips');
```

```python
#Grouping with stripplot()
sns.stripplot(x='sex',y='tip',hue='day',data=tips)
plt.ylabel("Tip($)");
```

```python
sns.stripplot(x='day',y='tip',hue='sex',data=tips)
plt.ylabel("Tip($)");
```

```python
sns.swarmplot(x='day',y='tip',hue='sex',data=tips)
plt.ylabel('Tip($)');
```

```python
#Cahnging Orientation
sns.swarmplot(x='tip',y='day',hue='sex',data=tips,orient='h')
plt.ylabel('Tip($)');
```

```python
#subplots
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('Tip($)')
plt.subplot(1,2,2)
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('Tip($)');
```
