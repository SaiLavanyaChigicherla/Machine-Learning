```python
data_cleaned1=data.drop_duplicates()
```

```python
data.shape
```

```python
data_cleaned1.shape
```

Rename Columns

```python
#Rename solar column
data_cleaned2=data_cleaned1.rename({"Solar.R":"Solar"},axis=1)
```

```python
data_cleaned2
```

Drop Columns

```python
data_cleaned2=data_cleaned2.drop('Year',axis=1)
```

```python
data_cleaned2.shape
```

Missing Values and Imputation

```python
data_cleaned2.isnull()
```

```python
data_cleaned2.notnull()
```

```python
data_cleaned2.isna()
```

```python
data_cleaned2.isna().any()
```

```python
data_cleaned2.isna().sum()
```

```python
data_cleaned2.info()
```

```python
#Mean Imputation
mean=data_cleaned2['Ozone'].mean()
mean
```

```python
data_cleaned2['Ozone']=data_cleaned2['Ozone'].fillna(mean)
```

```python
mean=data_cleaned2['Solar'].mean()
mean
```

```python
data_cleaned2['Solar']=data_cleaned2['Solar'].fillna(mean)
```

```python
data_cleaned2.isnull().sum()
#Plot Boxplot
```

```python
data_cleaned2['Weather']=data_cleaned2['Weather'].fillna(data_cleaned2['Weather'].mode().iloc[0])
```

```python
data_cleaned2['Month']=data_cleaned2['Month'].fillna(data_cleaned2['Month'].mode().iloc[0])
```
